# Cleaning Data

Tahap cleaning data dilakukan untuk memastikan dataset siap digunakan
dalam proses Exploratory Data Analysis (EDA).

```python
# menghapus data duplikat
df = df.drop_duplicates()
```

```python
# mengubah kolom date menjadi datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# mengecek ulang informasi dataset
df.info()
```

```python
# mengecek missing values kembali
df.isnull().sum()
```

```python
# mengecek data duplikat kembali
df.duplicated().sum()
```

Pada tahap cleaning data dilakukan beberapa proses untuk memastikan dataset siap digunakan pada tahap analisis.

Proses cleaning yang dilakukan meliputi:

- menghapus data duplikat menggunakan drop_duplicates(),
- mengubah tipe data kolom date dari object menjadi datetime,
- melakukan validasi ulang terhadap missing values dan duplicate data.

Hasil cleaning menunjukkan bahwa dataset telah siap digunakan pada tahap Exploratory Data Analysis (EDA).
